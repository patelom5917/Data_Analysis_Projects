```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Change data type
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df[df['released'].astype('str').str[0:4]!= df['year'].astype('str')]
```

```python
## here some data release date year and year value are different therefore we correct that

df['released_year'] = df['released'].astype('str').str[:4]
df['released_year']=df['released_year'].astype('int64')
df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending= False)
```

```python
df.drop_duplicates()
```

```python
## Convert object type column into numerical
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_numeric = df.copy()
lb = LabelEncoder()
for column in df_numeric.columns:
    if (df_numeric[column].dtype == 'object'):
        df_numeric[column] = lb.fit_transform(df_numeric[column])
df_numeric
```

```python
df_numeric.corr()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df_numeric.corr(),annot=True)
plt.title('Correlation between different column')
plt.show()
```

```python
corr_pair = df_numeric.corr().unstack()
```

```python
corr_pair
```

```python
high_corr = corr_pair[corr_pair>0.5]
print(high_corr)
```
